```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import math
%matplotlib inline
```

```python
game_of_thrones_data = pd.read_csv('https://query.data.world/s/347kixktojui5tetmswnydj426qkgv')
```

```python
game_of_thrones_data.head(5)
```

```python
game_of_thrones_data.isnull()
```

```python
game_of_thrones_data.isnull().sum()
```

```python
#dropping all the null data
game_of_thrones_data.dropna(inplace=True)
```

```python
game_of_thrones_data.isnull().sum()
```

```python
#we will predict death year based on Book  of Death
#dropping unneccesary data
game_of_thrones_data.drop(['Name','Allegiances','Death Chapter','Book Intro Chapter','Gender','Nobility','CoK','SoS','FfC','DwD'],axis=1,inplace=True)
```

```python
game_of_thrones_data.head(10)
```

```python
#setting up dependent and independent variable
X = game_of_thrones_data.drop("Death Year" , axis=1)
y = game_of_thrones_data["Death Year"]


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
from sklearn.linear_model import LogisticRegression
logModel = LogisticRegression()

#fitting the data
logModel.fit(X_train,y_train)
predictions = logModel.predict(X_test)
```

# checking the accuracy of our model

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error
r2_score = logModel.score(X,y)
print(r2_score)
```
